```python
import pandas as pd
import numpy as np

dir = '../../data/breast'

def load_data(dir):
    org_st_count = pd.read_csv(f'{dir}/Out_gene_expressions_10000genes.csv', index_col=0)
    sc_exp = pd.read_table(f'{dir}/raw_somatosensory_sc_exp.txt', index_col=0)
    sc_anno = pd.read_table(f'{dir}/somatosensory_sc_labels.txt', header=None)
    st_location = pd.read_csv(f'{dir}/Out_rect_locations.csv', index_col=0)
    return org_st_count, sc_exp, sc_anno, st_location
```

```python
org_st_count, sc_exp, sc_anno, st_location = load_data(dir)
```
