```python
import pandas as pd
import mygene

# Load your single-cell expression matrix
sc_exp = pd.read_table('/home/lytq/Spatial-Transcriptomics-Benchmark/data/lymph/raw_somatosensory_sc_exp.txt', index_col=0)

# Extract the Ensembl gene IDs
ensembl_ids = sc_exp.index.tolist()

# Initialize mygene
mg = mygene.MyGeneInfo()

# Query mygene (you can increase `step` or use `as_dataframe=True` if preferred)
print("Converting Ensembl IDs to gene symbols...")
query_results = mg.querymany(ensembl_ids, scopes='ensembl.gene', fields='symbol', species='human', as_dataframe=True)

# Clean the result: drop rows without a symbol
query_results = query_results[~query_results['notfound']]
query_results = query_results[~query_results['symbol'].isnull()]

# Map Ensembl ID to gene symbol
ensembl_to_symbol = query_results['symbol'].to_dict()

# Filter sc_exp to genes that could be mapped
mapped_genes = [gene for gene in sc_exp.index if gene in ensembl_to_symbol]
sc_exp_mapped = sc_exp.loc[mapped_genes].copy()

# Replace rownames (index) with gene symbols
sc_exp_mapped.index = [ensembl_to_symbol[gene] for gene in sc_exp_mapped.index]

print("✅ Converted expression matrix:")
print(sc_exp_mapped.head())
```
