```python
import sys
sys.path.append('/home/lytq/Spatial-Transcriptomics-Benchmark/utils')
from load_st_data import load_mHypothalamus

import numpy as np
import pandas as pd
import scanpy as sc
import matplotlib.pyplot as plt
import seaborn as sns
import glob
import os

import warnings
warnings.filterwarnings('ignore')
```

```python
data_path = '/home/lytq/Spatial-Transcriptomics-Benchmark/data/mHypothalamus'
out_path = '/home/lytq/Spatial-Transcriptomics-Benchmark/Results/MERFISH/mHypothalamus/GroundTruth'
sample_list = ['-0.04', '-0.09', '-0.14', '-0.19', '-0.24']
```

```python
for sample in sample_list:
    adata = load_mHypothalamus(data_path, sample)
    
    save_path = os.path.join(out_path, sample)
    if not os.path.exists(save_path):
        os.makedirs(save_path)
    
    fig, axes = plt.subplots(1, 1, figsize=(6, 6))
    sc.pl.spatial(adata, color='layer_guess', ax=axes, show=False, spot_size=20)
    axes.set_title(f'Ground Truth ({sample})')
    axes.axis('off')
    axes.set_aspect('equal', adjustable='datalim')
    plt.tight_layout()
    plt.savefig(os.path.join(save_path, 'gt.pdf'), format='pdf', bbox_inches='tight')
    plt.savefig(os.path.join(save_path, 'gt.png'), dpi=300, bbox_inches='tight')
    plt.show()
```
